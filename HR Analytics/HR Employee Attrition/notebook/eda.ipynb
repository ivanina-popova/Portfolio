```python
# Handling duplicates
df.duplicated().sum()
```

```python
# Check for unique values
def checkunique(data):
    nunique = data.apply(lambda col : col.nunique())
    unique = data.apply(lambda col : col.unique())
    result= pd.DataFrame(
        {
             'N-Unique' : nunique,
             'Unique' : unique
        }
    )
    return result
```

```python
# Check for unique values
checkunique(df)
```

```python
# Removing unnecessary columns
df= df.drop(['EmployeeCount','Over18','StandardHours', 'EmployeeNumber','PerformanceRating'], axis=1)
df
```

```python
# Checking for outliers
plt.figure(figsize=(16,10))
plt.subplot(1,2,1)
sns.scatterplot(df['MonthlyIncome'])
plt.subplot(1,2,2)
sns.boxplot(df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Deleting outliers with IQR
Q1= df['MonthlyIncome'].quantile(0.25)
Q3= df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
clean_df=df[(df['MonthlyIncome']>=lower) & (df['MonthlyIncome']<=upper)]

clean_df.reset_index()
```

```python
# Post-cleaning validation
plt.figure(figsize=(16,10))
plt.subplot(1,2,1)
sns.scatterplot(clean_df['MonthlyIncome'])
plt.subplot(1,2,2)
sns.boxplot(clean_df)
plt.xticks(rotation=90)
plt.show()

print(f'Number of Rows Deleted: {df.shape[0] - clean_df.shape[0]}')
```

```python
# Save the clean dataset
clean_df.to_csv('../data/clean_hr_employee_attrition.csv', index=False)
```
